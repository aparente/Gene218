```python
F1=open('Shakespeare.txt',mode='rU')
D1={} ##keys are words, values are how many times they appear
for L1 in F1:
    L2=L1.strip().replace('.',' ').replace('\t','').replace(',',' ')
    L3=L2.split()  ##splits each line into words
    for w1 in L3:
        if w1 !='':
            if w1 in D1:
                D1[w1]=D1[w1]+1
            else:
                D1[w1]=1
i=0
for d1 in D1:  ##iterates through dictionary key by key
    #print d1, D1[d1]
    i+=1
    if i > 100:
        break
```

```python
##make a list of the keys
def MySortFunction1(w1):
    return(D1[w1])  ##input values are words, output values are D1 (number of times the word is used)
##can use return(-D1[w1]) if you want it to be sorted from most used to least used
Keylist1=D1.keys()
Keylist2=sorted(Keylist1, key=MySortFunction1)  #default is alphabetical sort

##can now go through keylist2, which is ordered compared to dictionary
i=0
for d1 in Keylist2:  ##iterates through dictionary key by key
    #print d1, D1[d1]
    i+=1
    if i > 100:
        break
```
